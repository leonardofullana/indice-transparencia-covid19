### Importando as bibliotecas necessárias

```python
import pandas as pd
import bar_chart_race as bcr
import numpy as np
```

### Lendo a base de dados completa, disponível no Google Sheets

```python
df = pd.read_csv('http://docs.google.com/spreadsheets/d/18z8H9sR13HDQ007kCKJRJ_Y9gNCacjHRPW9j28aRlUk/export?format=csv&gid=106766150')
df.head()
```

### Lidando com dados ausentes e ausência de nome das colunas

```python
df1 = df.replace(np.nan, 'Estado', regex=True)
df1.columns = df1.iloc[0]
df1.head()
```

### Eliminando as colunas desnecessárias

```python
df2 = df1.drop(columns = ['14/05 - 21/05', '03/04 - 09/04', '09/04 - 16/04', '16/04 - 23/04', '23/04 - 30/04', '30/04 - 07/05', '07/05 - 14/05', '21/05 - 28/05', 'Acumulada'])
df2.head()
```

### Transpondo as colunas para linhas, para virar um dataframe compatível com a biblioteca

```python
df_t = df2.T
df_t.columns = df_t.iloc[0]
df_t.head(12)
```

### Alterando o formato das datas para poder ser convertido em datetime ser legível pela biblioteca

```python
df_data = df_t.replace({'03/04': '2020-04-03', '09/04': '2020-04-09','16/04': '2020-04-16', 
              '23/04': '2020-04-23', '30/04': '2020-04-30', '07/05': '2020-05-07',
              '14/05': '2020-05-14', '21/05': '2020-05-21', '28/05': '2020-05-28'})
df_data2 = df_data.rename(columns={'Estado': 'date'})
df_data2.head()
```

### Transformando as datas em formato datetime

```python
df_data3 = df_data2.drop(index = 'Estado')
df_data3['date'] = df_data3['date'].astype('datetime64[ns]')
df_data3.head()
```

### Transformando a primeira coluna, com as datas corretas, no índice e transformando as outras colunas com valores numerais em integer (número)

```python
df_data_clean = df_data3.set_index('date')
df_data_clean.index.name = 'date'
df_data_clean2 = df_data_clean.astype(int)
df_data_clean.head()
```

### Verificando o formato dos dados

```python
print(df_data_clean2.dtypes) 
```

### Criando o gráfico de Bar Chart Race

```python
def summary(values, ranks):
    media = int(values.mean())
    s = f'Nota Média = {media:,.0f}'
    return {'x': .95, 'y': .10, 's': s, 'ha': 'right', 'size': 8}

bcr.bar_chart_race(df_data_clean2, period_summary_func=summary, steps_per_period=60, period_length=4000, 
                   title='Índice de transparência da Covid-19',
                   shared_fontdict={'family': 'Helvetica', 'weight': 'bold',
                                    'color': 'rebeccapurple'},
                  figsize=(10, 6))
```

### Informações relevantes sobre o gráfico
